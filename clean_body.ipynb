```python
import re
from ming import Session, create_datastore, datastore
from ming import Document, Field, schema
import goslate
from email.header import decode_header

def translate_to_en(text):
    gs = goslate.Goslate()
    t = gs.translate(text, 'en')
    return t

def clean_qp(temp):
    temp = re.sub('\*\*', unicode("",'utf8'), temp)
    return temp

def clean_sign_and_transf(temp, subject):
    
    ##Look if the message contains transfered message & delete the header
    temp = ('<start>\n' + temp + '\n<end>')
    has_trans = re.search('(\n-----(.*?)-----\n)', temp, re.DOTALL)
    
    if has_trans !=None:
        temp = re.sub ('(\n-----(.*?)\n\n)', '<deleted transfered header>\n', temp, flags = re.DOTALL)
    else:
        is_trans = re.match ('(Fw|Fwd):', subject, flags = re.IGNORECASE)
        if is_trans != None:
            has_trans = reg_h.findall (temp, re.MULTILINE)
            if has_trans !=[]:
                temp = temp.replace(has_trans[0][0], '<deleted transfered header>\n')
    
    ##Delete all the redundant responses to e-mail
    reg_h = "(\n>?\s?\*?\s?(From|De)\s?:)"
    reg_h = re.compile(reg_h, re.IGNORECASE)
    has_header = reg_h.findall (temp, re.MULTILINE)
    if has_header != []:
        temp = re.sub ('(\n>?\s?\*?\s?(From|De)\s?:.*?\<end\>)', '\n<deleted redundant message>\n<end>', temp, flags = re.DOTALL)
    temp = re.sub ('(\nOn.*?wrote:\n.*?<end>)', '\n<deleted redundant message>\n<end>', temp, flags = re.DOTALL)
    temp = re.sub('(\n(january|february|march|april|may|june|july|august|september|october|november|december|2013|2012|2014).*?<.*?@.*?>.*?\n.*?<end>)','\n<deleted redundant message>\n<end>', temp, flags = re.DOTALL|re.IGNORECASE)
    temp = re.sub ('(\nLe\s.*?crit\s?:\n.*?<end>)', '\n<deleted redundant message>\n<end>', temp, flags = re.DOTALL)
    temp = re.sub('[\S]{40,50000}', '<useless>', temp)
    ##Delete all signature in the rest of the mail:
    #retrieve list of names in the Person table for signature detection
    list_of_signature = []
    people = Person.query.find().all()
    for p in people:
        sign = p.first_name + ' ' + p.last_name
        sign = re.sub('\(', '', sign)
        sign = re.sub('\)', '', sign)
        list_of_signature.append(sign)
        sign = p.last_name + ' ' + p.first_name
        sign = re.sub('\(', '', sign)
        sign = re.sub('\)', '', sign)
        list_of_signature.append(sign)
    regx1 = "("+'|'.join(list_of_signature)+")"
    regx1 = "(\n\s?\*?\s?" + regx1 + ")"
    reg1 = re.compile(regx1, re.IGNORECASE)
    found = reg1.findall(temp)
    ##delete the signature if there is a match
    if found != []:
        temp = re.sub(found[0][0] + '.*?<deleted transfered header>', '\n<deleted signature>\n', temp, flags = re.DOTALL)
        temp = re.sub(found[0][0] + '.*?<end>', '\n<deleted signature>\n<end>', temp, flags = re.DOTALL)
    
    ##Delete the greatings and small signatures:
    temp = re.sub('\n(OK|ok|Ok)\n', 'I am ok with this', temp, flags = re.DOTALL)
    final = re.sub('\n\w+\s?\w+\s?[,\!\.]?\s?\n', '<\\>\n', temp, flags = re.DOTALL)
    final = re.sub('\n\w+\s?[,\!\.]?\s?\n', '<\\>\n', final, flags = re.DOTALL)
    final = re.sub('\n\n', '\n', final)
    final = re.sub('\n>+', '\n', final)
    final = re.sub('\n', ' ', final)
    return final

def clean_body(temp):
    qp = clean_qp(temp.body)
    clean = clean_sign_and_transf(qp, temp.subject)
    return clean

def clean_persons():
    persons = []
    for p in Person.query.find().all():
        fname = re.sub('\'', '',p.first_name)
        fname = re.sub('/', '',fname)
        fname = re.sub(' ', '',fname)
        fname = re.sub('\n', '',fname)
        fname = re.sub('\t', '',fname)
        fname = re.sub('[^a-zA-Z0-9_]', '',fname)
        p.first_name = fname
        lname = re.sub('\'', '',p.last_name)
        lname = re.sub('/', '',lname)
        lname = re.sub(' ', '',lname)
        lname = re.sub('\n', '',lname)
        lname = re.sub('\t', '',lname)
        lname = re.sub('[^a-zA-Z0-9_]', '',lname)
        p.last_name = lname
        persons.append(fname + ' ' + lname)
    odm_session.flush()
    return True
```

```python
from ming import Session, create_datastore, datastore
from ming import Document, Field, schema
from ming.odm import ODMSession
from ming.odm.declarative import MappedClass
from ming.odm.property import FieldProperty, RelationProperty, ForeignIdProperty
from datetime import datetime

bind = create_datastore('aimep')
session = Session(bind)
odm_session = ODMSession (doc_session=session)

class Category(MappedClass):

    class __mongometa__:
        session = odm_session
        name = 'category'

    _id = FieldProperty(schema.ObjectId)
    name = FieldProperty(str)
    
    emails = RelationProperty('Email')
    persons = RelationProperty('JoinCategoryPerson')
    
class Person(MappedClass):
    class __mongometa__:
        session = odm_session
        name = 'person'
        
    _id = FieldProperty(schema.ObjectId)
    email_adress = FieldProperty(str)
    first_name = FieldProperty(str)
    last_name = FieldProperty(str)
    signature = FieldProperty(str)
    info = FieldProperty(str)
    
    email_received = RelationProperty('JoinEmailPersonTo')
    
class Email(MappedClass):
    
    class __mongometa__:
        session = odm_session
        name = 'email'
        
    _id = FieldProperty(schema.ObjectId)
    date = FieldProperty(datetime)
    subject = FieldProperty(str)
    trans_subject = FieldProperty(str)
    body = FieldProperty (str)
    thread = FieldProperty(str)
    language = FieldProperty(str)
    clean_body = FieldProperty(str)
    g_labels = FieldProperty([str])
    
    category_id = ForeignIdProperty(Category)
    category = RelationProperty(Category)
    sender_id = ForeignIdProperty(Person)
    sender = RelationProperty(Person)
    to = RelationProperty('JoinEmailPersonTo')
    sentences = RelationProperty('Sentence')
    attachments = RelationProperty('Attachment')
    
    def short_view(self):
        print "Subject: " + self.subject
        print "From: " + self.sender.email_adress
        print "To: "
        for p in self.to:
            print p.person.email_adress
    
    def get_to_adress(self):
        temp = []
        for p in self.to:
            try:
                temp.append(p.person.email_adress)
            except: pass
        return temp
            
    def trans_body(self):
        trans_body = ''
        for s in self.sentences:
            try:
                trans_body = trans_body + ' ' + s.en_sentence
            except:
                trans_body = trans_body
        return trans_body
    
    def relevant_body(self):
        rel_body = ''
        for s in self.sentences:
            if s.pred_tag != 'n':
                try:
                    rel_body = rel_body +' ' + s.en_sentence
                except:
                    rel_body = rel_body
        return rel_body
    
class JoinCategoryPerson(MappedClass):
    class __mongometa__:
        session = odm_session
        name = 'joincategoryperson'
        
    _id = FieldProperty(schema.ObjectId)
    amount = FieldProperty(int)
    category_id = ForeignIdProperty(Category)
    category = RelationProperty(Category)
    person_id = ForeignIdProperty(Person)
    person = RelationProperty(Person)
    
class JoinEmailPersonTo(MappedClass):
    class __mongometa__:
        session = odm_session
        name = 'joinemailpersonto'
        
    _id = FieldProperty(schema.ObjectId)
    email_id = ForeignIdProperty(Email)
    email = RelationProperty(Email)
    person_id = ForeignIdProperty(Person)
    person = RelationProperty(Person)  
    
class Sentence(MappedClass):
    
    class __mongometa__:
        session = odm_session
        name = 'sentence'
        
    _id = FieldProperty(schema.ObjectId)
    sentence_string = FieldProperty(str)
    en_sentence = FieldProperty(str)
    tag = FieldProperty (str)
    pred_tag = FieldProperty (str)
    
    email_id = ForeignIdProperty(Email)
    email = RelationProperty(Email)
    

class Attachment(MappedClass):
    class __mongometa__:
        session = odm_session
        name = 'attachment'
        
    _id = FieldProperty(schema.ObjectId)
    name = FieldProperty(str)
    extension = FieldProperty(str)   
    
    email_id = ForeignIdProperty(Email)
    email = RelationProperty(Email)
    
class Tag_Sent (MappedClass):
    
    class __mongometa__:
        session = odm_session
        name = 'tag_sent'
    
    _id = FieldProperty(schema.ObjectId)
    lang = FieldProperty(str)
    sentence = FieldProperty(str)
    trans_sentence = FieldProperty(str)
    tag = FieldProperty(str)

print "DB correctly created/syncronized"
```

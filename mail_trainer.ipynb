```python
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer         
from nltk.stem import WordNetLemmatizer
import numpy as np
import sys
import re
from sklearn.cross_validation import ShuffleSplit
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from collections import defaultdict
from sklearn.metrics import classification_report
from sklearn.feature_extraction import DictVectorizer
EMAIL_OWNER = 'thierry.noesen@belvas.be'
corpus_BOW = []
corpus_subject = []
persons = []
companies = []
threads = []
tags = []
for m in Email.query.find().all():
    if m.category.name != 'to_classify':
        tag = m.category.name
        corpus_BOW.append(m.trans_body() + m.trans_subject)
        corpus_subject.append(m.trans_subject)
        persons_temp=[]
        if m.sender.email_adress != EMAIL_OWNER:
            persons_temp.append(m.sender.email_adress)
        for e in m.get_to_adress():
            if e != EMAIL_OWNER:
                persons_temp.append(e)
        persons.append(persons_temp)
        #company = re.findall('(@[\w\-][\w\-\.]+[a-zA-Z]{1,4})', persons_temp[0])[0]
        #companies.append(company)
        threads.append(m.thread)
        tags.append(tag)
import pandas as pd
d = pd.DataFrame(tags, columns=['tags'])
print d.tags.value_counts()/float(d.shape[0])
```

### Prediction

```python
class LemmaTokenizer(object):
    def __init__(self):
        self.wnl = WordNetLemmatizer()
    def __call__(self, doc):
        return [self.wnl.lemmatize(t) for t in nltk.word_tokenize(doc)]

vectoriser = TfidfVectorizer(tokenizer = LemmaTokenizer(), ngram_range =(1,3))
X_corpus = vectoriser.fit_transform(corpus_BOW)
X_subject = vectoriser.fit_transform(corpus_subject)
y = np.array(tags)

from sklearn import metrics
cv = ShuffleSplit(len(corpus_BOW), n_iter=1, test_size=0.2, random_state=0)
for cv_index, (train, test) in enumerate(cv):
    #innerant models for projections
    model_corpus = SVC(kernel = 'linear', C = 1, gamma = 0.01,  tol=0.0001, probability = True).fit(X_corpus[train], y[train])
    model_subject = SVC(kernel = 'linear', C = 1, gamma = 0.01,  tol=0.0001, probability = True).fit(X_subject[train], y[train])
    #levels of certainty for subject and body
    certainty_BOW = model_corpus.predict_proba(X_corpus)
    certainty_subject = model_subject.predict_proba(X_subject)
    n_rows, n_labels = certainty_BOW.shape
    #creating final matrix containing the information
    full_matrix = []
    categories = sorted(list(set(tags)),key=unicode.lower)
    for i in xrange(n_rows):
        row = {}
        for j in xrange(n_labels):
            row['B_'+ categories[j]] = certainty_BOW[i,j]
            row['S_'+ categories[j]] = certainty_subject[i,j]
            row['thread'] = threads[i]
            #row['company'] = companies[i]
            #row['person'] = senders[i]
        #row['person'] = persons[i][0]if persons[i]!=[] else ""
        for x in persons[i]:
            row[x] = 1
        full_matrix.append(row)
    vec = DictVectorizer()
    X_full = vec.fit_transform(full_matrix)
    #creating the model
    model_full = SVC(kernel = 'linear', C = 1, gamma = 0.01,  tol=0.0001, probability = True).fit(X_full[train], y[train])
    #Predictions models
    prediction_full = model_full.predict(X_full[test])
    prediction_BOW = model_corpus.predict(X_corpus[test])
    prediction_subject = model_subject.predict(X_subject[test])
    true = y[test]
    print "Classification only based on BOW"
    print classification_report (true, prediction_BOW)
    print "Classification only based on Subject"
    print classification_report (true, prediction_subject)
    print "Classification based on everything"
    print classification_report (true, prediction_full)
    #Looking for really behind numbers
    certainty_full = model_full.predict_proba(X_full[test])
    n_rows, n_labels = certainty_full.shape
    predictions_values = np.amax(certainty_full, axis = 1)
    predictions = np.argmax(certainty_full, axis = 1)
    pred_categories = []
    for p in predictions:
        pred_categories.append (categories[p])
    #building a verification matrix:
    verif_matrix = []
    for i in xrange(n_rows):
        row = {}
        for j in xrange(n_labels):
            row['certainty'] = predictions_values[i]
            row['prediction'] = pred_categories[i]
            row['reality'] = true[i]
            row['number']=test[i]
        verif_matrix.append(row)
    print "Average certainty: " + str(sum(predictions_values)/len(predictions_values))
    mod_matrix = verif_matrix
    print "____________________________"
    for i in mod_matrix:
        if i['prediction'] == i['reality']:
            bad_certainties.append(i['certainty'])
            print i['prediction'], certainty_full[mod_matrix.index(i),:], i['reality']
    bad_certainties = []
    print "____________________________"
    for i in mod_matrix:
        if i['prediction'] != i['reality']:
            bad_certainties.append(i['certainty'])
            print threads[i['number']], corpus_subject[i['number']], tags[i['number']]
            print i['prediction'], certainty_full[mod_matrix.index(i),:], i['reality']
    print "Average certainty for bad results: " + str(sum(bad_certainties)/len(bad_certainties))
```
